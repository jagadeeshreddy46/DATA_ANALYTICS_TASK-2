```python
import pandas as pd
df = pd.read_csv(r"C:\Users\mekal\Downloads\Global Store Sales_Dataset.csv", encoding="latin1")
df.head(3)
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'],format='%d-%m-%Y')
df['Ship Date']=pd.to_datetime(df['Ship Date'],format='%d-%m-%Y')
df.drop("Postal Code", axis=1, inplace=True)
```

```python
print("Shape",df.shape)
print("Information of Dataset",df.info())
print("No of Missing Values:",df.isnull().sum())
```

```python
#Histogram

#-----------------------Sales Distribution------------------

import matplotlib.pyplot as plt
import seaborn
import seaborn as sns
plt.figure(figsize=(10,6))
df['Sales'].hist()
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.title("Sales Distribution")
plt.show()
```

```python
#Profit Distribution
plt.figure(figsize=(10,6))  
df['Profit'].hist(color='red', alpha=0.7 )
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.title("Profit Distribution")
plt.show()
```

```python
#Category Distribution

df['Category'].value_counts().plot(kind='bar',color='orange', figsize=(10,6),fontsize=10)
plt.xlabel("Category")
plt.ylabel("Count")
plt.title("Category Distribution")
plt.show()
```

```python
#Sales Box Plot
plt.figure(figsize=(10,6))
plt.boxplot(df['Sales'])
plt.xlabel("Sales")
plt.ylabel("Value")
plt.title("Sales Box Plot")
plt.show()
```

```python
#Scatter Plot for Sales Vs Profit
plt.figure(figsize=(10,6))
sns.scatterplot(x='Sales',y='Profit',data=df,hue='Category')
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales Vs Profit")
plt.show()
```

```python
#Discount Vs profit
plt.figure(figsize=(10,6))
sns.scatterplot(x='Discount',y='Profit',data=df,hue='Category')
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.title("Discount Vs Profit")
plt.show()
```

```python
#Sub-Category Distribution
df['Sub-Category'].value_counts().plot(kind='bar', figsize=(12,6),color='green',fontsize=10)
plt.xlabel("Sub-Category")
plt.ylabel("Count")
plt.title("Sub-Category Distribution")
plt.show()
```

```python
df.groupby('Category')['Sales'].sum().plot(kind='bar', figsize=(10,6),color='purple',fontsize=10    )
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.title("Total sales by Category")
plt.show()
```

```python
import plotly.express as px

df['YearMonth'] = df['Order Date'].dt.to_period('M').astype(str)

monthly_sales = df.groupby('YearMonth')['Sales'].sum().reset_index()

fig = px.line(monthly_sales, 
              x='YearMonth', 
              y='Sales',
              title="Monthly Sales Trend",
              markers=True)

fig.show()
```

```python
import plotly.express as px

fig = px.histogram(df,
                   x="Sales",
                   nbins=50,
                   title="Sales Distribution")  # adds small boxplot on top

fig.show()
```

```python
#Time Series Analysis
plt.figure(figsize=(12,6))
df['YearMonth'] = df['Order Date'].dt.to_period('M')
df.groupby('YearMonth')['Sales'].sum().plot(marker='o')
plt.xlabel("Year-Month")
plt.ylabel("Total Sales")
plt.title("Monthly Sales Trend")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

loss_df = df[df['Profit'] < 0]

category_loss = loss_df.groupby('Category')['Profit'].sum()

plt.figure(figsize=(8,5))
category_loss.plot(kind='bar', color='red')

plt.xlabel("Category")
plt.ylabel("Total Loss")
plt.title("Total Loss by Category")

plt.axhline(0, color='black')  # baseline
plt.show()
```

```python
#High Discount Analysis

high_discount=df[df['Discount']>0.5]
high_discount.groupby('Category')['Sales'].sum().plot(kind='bar',color='orange',fontsize=10,figsize=(10,6))
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.title("Total Sales for High Discount")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[['Sales','Quantity','Discount','Profit','Shipping Cost']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Advanced Interactive Scatter Plot
import plotly.express as px
fig=px.scatter(df, x='Sales', y='Profit', color='Category', title="Sales vs Profit by Category")
fig.show()
```

```python
import matplotlib.pyplot as plt

segment_sales = df.groupby('Segment')['Sales'].sum()

plt.figure(figsize=(5,7))

plt.pie(segment_sales,
        labels=segment_sales.index,
        autopct='%1.1f%%',
        startangle=90,
        shadow=True)

plt.title("Sales Contribution by Segment", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
pivot = df.pivot_table(values='Sales',
                       index='Region',
                       columns='Category',
                       aggfunc='sum')

pivot.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Region vs Category Sales")
plt.ylabel("Sales")
plt.show()
```
